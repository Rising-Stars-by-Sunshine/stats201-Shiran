## **Question 1**
*Did you get the same data querying via different node operators?*

The following code outputs a table of Infura, Alchemy, and QuickNode data query results of Ethereum before and after the merge, with a query radius of 10. As seen, the results of the three tables are consistent with one another, and therefore the same data are indeed achieved.

This is also consistent with theory. Since decentralized consensus is reached in blockchain systems, different nodes should have received and recorded the same data.

```python
import pandas as pd
pd.set_option('display.max_rows',None)

print('Infura Node')
!ethereumetl export_blocks_and_transactions --start-block 15537383 --end-block 15537403 --blocks-output blocks1.csv --provider-uri https://mainnet.infura.io/v3/fbada49129ac43f7b9d141e90f0359e6
display(pd.read_csv('blocks1.csv'))
print('Alchemy Node')
!ethereumetl export_blocks_and_transactions --start-block 15537383 --end-block 15537403 --blocks-output blocks2.csv --provider-uri https://eth-mainnet.g.alchemy.com/v2/0UiZNJos8ZxD_wnB9fPOIq6l_4gI_8VK
display(pd.read_csv('blocks2.csv'))
print('QuickNode Node')
!ethereumetl export_blocks_and_transactions --start-block 15537383 --end-block 15537403 --blocks-output blocks3.csv --provider-uri https://blissful-compatible-smoke.discover.quiknode.pro/148b0508b981bde3e1823f171e7895fe7e7e9fb0/
display(pd.read_csv('blocks3.csv'))
```
